```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
categories = np.array([['red'], ['green'], ['blue'], ['green'], ['red']])

encoder = OneHotEncoder(sparse_output=False)

onehot_encoded = encoder.fit_transform(categories)
```

```python
print('Categories:', encoder.categories_)

print('One-hot encoded array:')
print(onehot_encoded)
```

```python
original_labels = encoder.inverse_transform(onehot_encoded)
print('Recovered original labels:')
print(original_labels)
```

```python
raw_data = [['apple'], ['banana'], ['apple'], ['orange']]
encoder2 = OneHotEncoder(sparse_output=False)
encoded_data = encoder2.fit_transform(raw_data)
print('\nRaw data:', raw_data)
print('Categories learned:', encoder2.categories_)
print('One-hot encoded data:')
print(encoded_data)
```
